```python
import numpy as np
def grad(x):
    return 2*x+ 5*np.cos(x) # Derivative of the cost function
def cost(x):
    return x**2 + 5*np.sin(x) # Cost function

def myGD1(x0, eta):
    x = [x0]
    for it in range(100):
        x_new = x[-1] - eta*grad(x[-1])    # Following 12.2
        if abs(grad(x_new)) < 1e-3: # just a small number
            print(f"{it} Iteration:{grad(x_new)} < 1e-3 | {x_new}")
            print('--Converged--')
            break
        elif it % 5 == 0:
            print(f"{it} Iteration:{grad(x_new)} | {x_new}")
        x.append(x_new)
    return (x, it)
```

```python
(x1, it1) = myGD1(-5, .1)
(x2, it2) = myGD1(5, .1)
```
